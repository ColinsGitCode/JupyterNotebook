```python
def DictAppend(dic,name):
    if name in dic.keys():
        # dic[name].value = dic[name].value + 1
        dic[name] = dic[name] + 1
    else:
        dic.setdefault(name,1)
```

```python

def GetEachPacketDnsQueryName(Cap):
    distinct_name ={}
    def EachNameProcess(*args):
        objlist = dir(args[0])
        if 'dns' in objlist:
            the_name = args[0].dns.qry_name
            if the_name != " ":
                DictAppend(distinct_name,the_name)
            else:
                pass
        else:
            pass
        
    Cap.apply_on_packets(EachNameProcess, timeout=10000)
    return distinct_name
    
DistinctName = GetEachPacketDnsQueryName(DNS)
print(DistinctName)
```
