```python
#常用的引用
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

# 1，利用数组进行数据处理

用数组表达式代替循环的做法，通常被称为矢量化(Vectorization)。一般来说，矢量化数组运算要比等价的纯Python方式快一至两个数量级，甚至更多

```python
# 1000个间距相同的点
points = np.arange(-5,5,0.01)
points
```

```python
# np.meshgrid()函数接受两个一维数组，并产生两个二维矩阵（对应于两个数组中所有的（x,y）对）
xs,ys = np.meshgrid(points,points)
xs
```

```python
# 之后可以进行矢量化情况下的两个多维数组的计算和其他操作
# 实例如下：
Z = np.sqrt(xs**2 + ys**2)
# 利用matplotlib库进行绘图
plt.imshow(Z, cmap=plt.cm.gray); plt.colorbar()
plt.title("Image polt of $\sqrt{x^2 + y^2}$ for a grid of values")
```
